```python
import pandas as pd
import matplotlib.pyplot as plt
import requests

```

```python
# resibe datos de la api y los transforma a formato json
response = requests.get("https://api.gael.cloud/general/public/clima/")
data = response.json()
```

```python
# Filtrar los datos para eliminar la estación en la Península de Antártida
filtered_data = [d for d in data if d["Estacion"] != "Península Antártica"]
```

```python


estaciones = [d["Estacion"] for d in filtered_data]
temperaturas = [float(d["Temp"]) for d in filtered_data]

plt.figure(figsize=(10, 6))
bars = plt.bar(estaciones, temperaturas)
plt.xlabel('Estación')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas por Zonas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Ejemplo 2: Gráfico de dispersión de temperatura vs. humedad
temperaturas = [float(d["Temp"]) for d in data]
humedades = [int(d["Humedad"]) if d["Humedad"].isdigit() else None for d in data]

# Filtra valores None (valores faltantes) para evitar el error
temperaturas = [t for t, h in zip(temperaturas, humedades) if h is not None]
humedades = [h for h in humedades if h is not None]

plt.scatter(temperaturas, humedades)
plt.xlabel('Temperatura (°C)')
plt.ylabel('Humedad (%)')
plt.title('Temperatura vs. Humedad')
plt.grid(True)
plt.show()





```
