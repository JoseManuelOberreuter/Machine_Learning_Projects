# APLICACIÓN DE ALGORITMOS DE MACHINE LEARNING

```python
#IMPORTAMOS LIBRERIAS
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#SEPARAMOS LOS DATOS CON LAS CARACTERISTICAS Y LAS ETIQUETAS O RESULTADOS

X = np.array(iris.drop('Species', axis=1))
y = np.array(iris['Species'])
```

```python
#SEPARAMOS LOS DATOS DE 'TRAIN' EN ENTRENAMIENTO Y PRUEBA PARA PROBRAR ALGORITMOS

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('Son {} datos para entrenamiento y {} datos para prueba'.format(X_train.shape[0], X_test.shape[0]))
```

```python
# MODELO DE REGRESION LOGICA

algoritmo = LogisticRegression(max_iter=1000)  
algoritmo.fit(X_train, y_train)  

# Realizar predicciones en los datos de prueba
Y_pred = algoritmo.predict(X_test)  

# Evaluar la precisión en los datos de prueba
print('Precisión Regresión Logística: {}'.format(algoritmo.score(X_test, y_test)))  
```

```python
# MODELO DE MÁQUINAS DE VECTORES DE SOPORTE

algoritmo = SVC()
algoritmo.fit(X_train, y_train)

# Realizar predicciones en los datos de prueba
Y_pred = algoritmo.predict(X_test)

print('Precisión Máquinas de Vectores de Soporte: {}'.format(algoritmo.score(X_train, y_train)))
```

```python
# MODELOS DE VECINOS MÁS CERCANOS

# Definimos el n° de vecinos 
algoritmo = KNeighborsClassifier(n_neighbors=5)

#Entrenamos el algoritmo
algoritmo.fit(X_train, y_train)

# Realizar predicciones en los datos de prueba
Y_pred = algoritmo.predict(X_test)
print('Precisión Vecinos más Cercanos: {}'.format(algoritmo.score(X_train, y_train)))
```

```python
#MODELO DE ÁRBOLES DE DECISIÓN CLASIFICACIÓN
algoritmo = DecisionTreeClassifier()
algoritmo.fit(X_train, y_train)
Y_pred = algoritmo.predict(X_test)
print('Precisión Árboles de Decisión Clasificación: {}'.format(algoritmo.score(X_train, y_train)))
```

```python
# EJEMPLO 1: IRIS-SETOSA

nueva_flor_setosa = np.array([5.1, 3.5, 1.4, 0.2])  # Medidas típicas de Iris-setosa

clase_predicha_setosa = algoritmo.predict([nueva_flor_setosa])
print('Especie predicha para la nueva flor (Iris-setosa): {}'.format(clase_predicha_setosa[0]))
```
